```python
import matplotlib.pyplot as plt
from scipy import signal
```

```python
#constants
k = 10
m = 10
D = [4, 8, 16, 20, 40, 80]
```

```python
for i in range(0,5):
    A = [[0, 1], [-k/m, -D[i]/m]]
    B = [[0], [k/m]]
    C = [1, 0]
    E = [0]
    my_ss = signal.StateSpace(A, B, C, E)
    t, y = signal.step(my_ss)
    plt.plot(t, y)

plt.legend(['0.2', '0.4', '0.8', '1', '2', '4'])
```
