```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#constants
iterator = 800
h = 0.0002
a11 = -50
a21 = -19000
a22 = -21.5
ein = 1.5

v10 = 0.0
v20 = 0.0

t =np.zeros(shape=(iterator))
v1 =np.zeros(shape=(iterator))
v2 =np.zeros(shape=(iterator))
```

```python
def fun1(a11,v10,ein):
    return(a11*v10-a11*ein)

def fun2(a21,a22,v10,v20,ein):
    return(a21*v10+a22*v20-a21*ein)
```

```python
for i in range(1,iterator):
    m11 = fun1(a11,v10,ein)
    m12 = fun1(a11,v10+m11*h/2,ein)
    m13 = fun1(a11,v10+m12*h/2,ein)
    m14 = fun1(a11,v10+m13*h,ein)
    v11=v10+((m11+2*m12+2*m13+m14)/6)*h;
    m21=fun2(a21, a22, v10,v20,ein);
    m22=fun2(a21, a22, v10+h/2,v20+m21*h/2,ein);
    m23=fun2(a21, a22, v10+h/2,v20+m22*h/2,ein);
    m24=fun2(a21, a22, v10+h,v20+m23*h,ein);
    v21=v20+((m21+2*m22+2*m23+m24)/6)*h;
    v10=v11;
    v20=v21;
    t[i] = h*i;
    v1[i] = v10;
    v2[i] = v20;
```

```python
#As the range is uncomparable different plots are used for v1 and v2
fig = plt.figure()
plt.plot(v1,label ="v1")
plt.legend()
plt.show()
plt.plot(v2,label ="v2")
plt.legend()
plt.show()
```
