```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# constants 
iterations = 200
c_1 = np.ones(shape=(iterations))
c_2 = np.ones(shape=(iterations)) * 2
c_3 = np.zeros(shape=(iterations))
dt= 0.001
i = 0
```

```python
# Simulation
k1 = 3
k2 = 5
for t in range(1, iterations):
    c_1[t] = c_1[t - 1] + (k2 * c_3[t - 1] - k1 * c_1[t - 1] * c_2[t - 1]) * dt 
    c_2[t] = c_2[t - 1] + (k2 * c_3[t - 1] - k1 * c_1[t - 1] * c_2[t - 1]) * dt 
    c_3[t] = c_3[t - 1] + (2 * k1 * c_1[t - 1] * c_2[t - 1] - 2 * k2 * c_3[t - 1]) * dt 
```

```python
fig = plt.figure()
plt.plot(c_1, label = "c1")
plt.plot(c_2, label = "c2")
plt.plot(c_3, label = "c3")

plt.legend()
plt.show()
```
